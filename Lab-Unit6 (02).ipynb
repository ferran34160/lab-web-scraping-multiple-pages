```python
from random import randint
from time import sleep

iterations = range(2010, 2017, 1)

pages = []
for i in iterations:
    start_at= str(i)
    url = "https://playback.fm/charts/top-100-songs/"+ start_at +""

    # download html with a get request:
    response = requests.get(url)
    #response = requests.get(url, headers = {"Accept-Language": "en-US"})

    # monitor the process by printing the status code
    print("Status code: " + str(response.status_code))

    # store response into "pages" list
    pages.append(response)

    # respectful nap:
    wait_time = randint(1,4)
    print("I will sleep for " + str(wait_time) + " second/s.")
    sleep(wait_time)
```

```python
soup = BeautifulSoup(pages[0].content, "html.parser")
```

```python
soup.select("td:nth-child(2)")
str.strip(soup.select("a.artist")[1].get_text())
soup.select("td.mobile-hide > a > span.song")[0].get_text()
```

```python
pages_parsed = []
titles = []
artist = []

for i in range(len(pages)):
    # parse all pages
    pages_parsed.append(BeautifulSoup(pages[i].content, "html.parser"))
    
    songs_html = pages_parsed[i].select("td:nth-child(2)")
    
    for j in range(len(songs_html)):
        artist.append(songs_html[j].select("a.artist")[0].get_text().strip())
        titles.append(songs_html[j].select("span.song")[0].get_text().strip())
```

```python
# in 2013 and 2014 there are only 99 songs

import pandas as pd
top_100_2010_2016 = pd.DataFrame({"songs_title":titles,
                       "artists":artist,
                      })

top_100_2010_2016
```

List all language names and number of related articles in the order they appear in wikipedia.org: url = 'https://www.wikipedia.org/'

```python
url = "https://www.wikipedia.org/"
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup.select("div.central-featured-lang")[0]
```

```python
wikipedia_lang=[]
wikipedia_articles=[]
for language in soup.select("div.central-featured-lang"):
    wikipedia_lang.append(language.select("strong")[0].get_text())
    wikipedia_articles.append(language.select("bdi")[0].get_text().replace(u'\xa0', u' '))
    
wikipedia = pd.DataFrame({"language":wikipedia_lang,
                       "articles":wikipedia_articles})

wikipedia
```
